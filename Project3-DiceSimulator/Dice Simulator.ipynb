### Project 3: Dice Simulator

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Simulate rolling 2 dice 1,000 times
np.random.seed(42)
dice1 = np.random.randint(1,7,1000)
dice2 = np.random.randint(1,7,1000)
sums = dice1 + dice2
```

```python
# Step 2: Count occurrences of each sum (2â€“12)
unique, counts = np.unique(sums, return_counts= True)
sum_count= dict(zip(unique, counts))
print("Sum occurrences:\n", sum_count)
```

```python
# Step 3: Calculate probability of rolling a 7
prob_7 = sum_count.get(0,7) / 1000
print(f"\nProbability of rolling a 7: {prob_7:.2%}")
```

```python
# Step 4: Plot histogram of sums
plt.figure(figsize=(8,5))
plt.hist(sums, bins=np.arange(1.5, 12.5, 1), edgecolor='black', color='orange')
plt.xlabel("Dice Sum")
plt.ylabel("Frequency")
plt.title("Histogram of Dice Rolls (2 Dice, 1000 Trials)")
plt.show()
```
