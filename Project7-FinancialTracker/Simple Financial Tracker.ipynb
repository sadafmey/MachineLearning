### Simple Financial Tracker (Pandas) â€“ monthly income vs expenses

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Create a simple financial dataset
data = {
    "Month": [
        "January", "February", "March", 
        "April", "May", "June"
    ],
    "Income": [3000, 3200, 2800, 3500, 3600, 3300],
    "Expenses": [2500, 2700, 3000, 3200, 3100, 2900]
}
df = pd.DataFrame(data)
print("Financial Data:\n", df)
```

```python
# Step 2: Calculate net savings (Income - Expenses)
df["Savings"] = df["Income"] - df["Expenses"]
print("\nFinancial Data with Savings:\n", df)
```

```python
# Step 3: Plot Income vs Expenses
plt.figure(figsize=(10, 6))
plt.plot(df["Month"], df["Income"], marker='o', label="Income")
plt.plot(df["Month"], df["Expenses"], marker='o', label="Expenses")
plt.plot(df["Month"], df["Savings"], marker='o', linestyle='--', label="Savings")
plt.title("Monthly Income vs Expenses")
plt.xlabel("Month")
plt.ylabel("Amount ($)")
plt.legend()
plt.grid(True)
plt.show()
```
