### Project 2: Sales Data Analysis (Pandas)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Create a Pandas DataFrame
data = {
     'Product': ['Laptop', 'Phone', 'Tablet', 'Headphones', 'Smartwatch'],
    'Price': [1000, 500, 300, 50, 200],
    'Quantity_Sold': [5, 20, 10, 50, 15]
}

df = pd.DataFrame(data)
print("Sales DataFrame:\n", df)
```

```python
# Step 2: Compute total revenue for each product
df['Revenue'] = df['Price'] * df['Quantity_Sold']
print("\nDataFrame with Revenue:\n", df)
```

```python
# Step 3: Find the best-selling product by revenue
best_selling = df.loc[df['Revenue'].idxmax()]
print("\nBest-Selling Product:\n", best_selling)
```

```python
# Step 4: Plot a horizontal bar chart of sales revenue
plt.figure(figsize=(8, 5))
plt.barh(df['Product'], df['Revenue'], color='teal')
plt.xlabel('Revenue ($)')
plt.title('Sales Revenue by Product')
plt.gca().invert_yaxis()  # Highest revenue on top
plt.show()
```
