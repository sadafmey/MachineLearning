### Daily Step Counter (Pandas + Matplotlib) â€“ bar chart of steps per day

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Create daily steps dataset
data = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
    "Steps": [7500, 8200, 9000, 6500, 7000, 10000, 12000]
}
df = pd.DataFrame(data)
print("Daily Steps Data:\n", df)
```

```python
# Step 2: Plot a bar chart of steps per day
plt.figure(figsize=(10, 6))
plt.bar(df["Day"], df["Steps"], color='skyblue', edgecolor='black')
plt.title("Daily Steps Count")
plt.xlabel("Day")
plt.ylabel("Number of Steps")
plt.ylim(0, max(df["Steps"]) + 2000)  # Add some space above highest bar
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
