### Project 1: Student Scores Analysis (NumPy + Matplotlib)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# step 1: create numpy array of 10 students and 3 subjects with random scores (0,100)
np.random.seed(42) # for reproducibility 
scores = np.random.randint(0, 101, size=(10,3))
print("Scores Array:\n", scores)
```

```python
# Step 2: Calculate each student's average score
average_score = scores.mean(axis=1)
print("\nAverage Scores:\n", average_score)
```

```python
# Step 3: Find the top 3 students
top_3_students = average_score.argsort()[::-1][:3] # indices of top 3 students
print("\nTop 3 Students (indices):", top_3_students)
print("Top 3 Students (average scores):", average_score[top_3_students])
```

```python
# Step 4: Plot a bar chart of average scores
plt.figure(figsize=(10,6))
plt.bar(range(1,11), average_score, color="teal")
plt.title('Average score of 10 students')
plt.xlabel('Student Number')
plt.ylabel('Average Score')
plt.show()
```
